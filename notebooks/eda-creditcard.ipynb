```python
# ==========================================
# EDA - CREDIT CARD DATASET
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

credit = pd.read_csv("../data/raw/creditcard.csv")

# --------------------------
# Overview
# --------------------------

print(credit.shape)
print(credit.head())
print(credit.info())

# --------------------------
# Missing Values
# --------------------------

print(credit.isnull().sum())

# --------------------------
# Duplicates
# --------------------------

print(credit.duplicated().sum())

credit = credit.drop_duplicates()

# --------------------------
# Target Distribution
# --------------------------

print(credit["Class"].value_counts())
print(credit["Class"].value_counts(normalize=True)*100)

sns.countplot(data=credit, x="Class")
plt.title("Credit Card Fraud Distribution")
plt.show()

# --------------------------
# Amount Distribution
# --------------------------

sns.histplot(
    credit["Amount"],
    kde=True
)

plt.title("Transaction Amount")
plt.show()

# --------------------------
# Time Distribution
# --------------------------

sns.histplot(
    credit["Time"],
    kde=True
)

plt.title("Transaction Time")
plt.show()

# --------------------------
# Fraud vs Amount
# --------------------------

sns.boxplot(
    data=credit,
    x="Class",
    y="Amount"
)

plt.show()

# Correlation Heatmap

plt.figure(figsize=(14,10))

sns.heatmap(
    credit.corr(),
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()

print("EDA Completed")
```
