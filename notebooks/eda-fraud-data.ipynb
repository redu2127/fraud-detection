```python
# =====================================================
# FRAUD DETECTION PROJECT
# NOTEBOOK 1: DATA UNDERSTANDING & CLEANING
# =====================================================

# =====================
# IMPORT LIBRARIES
# =====================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

plt.style.use("ggplot")
%matplotlib inline

print("="*60)
print("LIBRARIES LOADED SUCCESSFULLY")
print("="*60)

# =====================
# LOAD DATA
# =====================

fraud_df = pd.read_csv("../data/raw/Fraud_Data.csv")
ip_df = pd.read_csv("../data/raw/IpAddress_to_Country.csv")

print("\nDATASETS LOADED SUCCESSFULLY")
print("-"*60)

print(f"Fraud Dataset Shape: {fraud_df.shape}")
print(f"IP Dataset Shape: {ip_df.shape}")

# =====================
# FIRST LOOK
# =====================

print("\nFIRST 5 ROWS")
print("-"*60)

display(fraud_df.head())

print("\nCOLUMN NAMES")
print("-"*60)

print(fraud_df.columns.tolist())

# =====================
# DATASET INFORMATION
# =====================

print("\nDATASET INFORMATION")
print("-"*60)

fraud_df.info()

# =====================
# SUMMARY STATISTICS
# =====================

print("\nSUMMARY STATISTICS")
print("-"*60)

display(fraud_df.describe(include="all").T)

# =====================
# MISSING VALUES
# =====================

print("\nMISSING VALUES ANALYSIS")
print("-"*60)

missing_values = pd.DataFrame({
    "Missing Count": fraud_df.isnull().sum(),
    "Missing Percentage": round(
        fraud_df.isnull().mean()*100,
        2
    )
})

display(
    missing_values.sort_values(
        by="Missing Percentage",
        ascending=False
    )
)

# =====================
# MISSING VALUE HEATMAP
# =====================

plt.figure(figsize=(12,6))

sns.heatmap(
    fraud_df.isnull(),
    cbar=False,
    yticklabels=False
)

plt.title("Missing Values Heatmap")
plt.show()

# =====================
# DUPLICATES
# =====================

print("\nDUPLICATE RECORD ANALYSIS")
print("-"*60)

duplicates = fraud_df.duplicated().sum()

print(f"Duplicate Rows Found: {duplicates}")

# =====================
# REMOVE DUPLICATES
# =====================

fraud_df = fraud_df.drop_duplicates()

print(f"Shape After Removing Duplicates: {fraud_df.shape}")

# =====================
# CONVERT DATETIME
# =====================

fraud_df["signup_time"] = pd.to_datetime(
    fraud_df["signup_time"]
)

fraud_df["purchase_time"] = pd.to_datetime(
    fraud_df["purchase_time"]
)

print("\nDATETIME CONVERSION COMPLETED")

# =====================
# TARGET DISTRIBUTION
# =====================

print("\nCLASS DISTRIBUTION")
print("-"*60)

display(
    fraud_df["class"].value_counts()
)

print("\nCLASS PERCENTAGE")
print("-"*60)

display(
    round(
        fraud_df["class"].value_counts(normalize=True)*100,
        2
    )
)

# =====================
# CLASS IMBALANCE PLOT
# =====================

plt.figure(figsize=(6,4))

sns.countplot(
    x="class",
    data=fraud_df
)

plt.title("Fraud vs Non-Fraud Transactions")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()

# =====================
# DATA TYPES CHECK
# =====================

print("\nDATA TYPES")
print("-"*60)

display(fraud_df.dtypes)

# =====================
# SAVE CLEANED DATA
# =====================

fraud_df.to_csv(
    "../data/processed/fraud_cleaned.csv",
    index=False
)

print("\nCleaned dataset saved successfully.")
print("Location: data/processed/fraud_cleaned.csv")

# =====================
# FINAL SUMMARY
# =====================

print("\n" + "="*60)
print("TASKS COMPLETED")
print("="*60)

print("✓ Dataset Loaded")
print("✓ Dataset Structure Examined")
print("✓ Missing Values Checked")
print("✓ Duplicate Records Removed")
print("✓ Datetime Columns Converted")
print("✓ Class Distribution Analyzed")
print("✓ Class Imbalance Visualized")
print("✓ Cleaned Dataset Saved")

print("="*60)
```
